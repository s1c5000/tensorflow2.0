## 6.1 특징추출

보스턴 주택 가격 데이터세트에는 주택가격을 예측하기위한 특징들이 들어있다. 이 특징들은 분류되어있기때문에 추출하기 쉬웠다.     
이미지데이터의 경우는 어떠할까?     
과거엔 이미지의 특성을찾기위한 다양한 방법이 개발되었다.   이러한 특징추출기법중 하나인 컨볼루션은 각 픽셀을 본래 픽셀과 그 주변 픽셀의 조합으로 대체하는 동작이다. 이때 연산에 쓰이는 작은 행렬을 filter나 kernel이라고 한다. 

컨볼루션은 우리말로 합성곱이라고한다. 원본 이미지의 각 픽셀을 포함한 주변 픽셀과 필터의 모든 픽셀은 각각곱연산을 하고, 그 결과를 모두합해서 새로운 이미지에 넣어주기 때문에 합성곱이라는 이름이 붙었다.

image.png

이러한 필터들은 경험적 지식을 통해 직접 손으로 값을 넣어준 결과이다. 이것을 수작업으로 설계한 특징(hand crafted feature)라고 한다.     
문제는 전문적인지식이 필요하고, 시간과 비용이 많이들고, 한 분야에 효과적인 특징을 다른 분야에 적용하기 어렵다.

딥러닝 기반의 컨볼류션연산은 이런 문제접들을 모두 해결했다. 네트워크가 특징을 추출하는 필터를 자동으로 생성한다.      
학습을 계속하면 네트워크를 구성하는 각 뉴런들은 입력한 데이터에 대해 특정 패턴을 잘 추출할 수있도록 적응하게 된다.

### 6.2.1 컨볼루션 레이어

컨볼루션레이어는 말 그대로 컨볼루션 연산을 하는 레이어이다. 다만 여기서 사요하는 필터는 사람이 미리정해놓은 것이 아니라 네트워크의 학습을 통해 자동으로 추출된다. 우리가 지정하는것은 필터의 개수 정도이다.

보통 컬러이미지는 RGB 채널을 갖고 흑백은 하나뿐이다     
컨볼루션 레이어는 각 채널에 대해 계산된 값을 합쳐서 새로운 이미지를 만들어낸다. 이때 새로운 이미지의 마지막 차원 수는 필터의 수와 동일하다. 일반적인 컨볼루션 신경망은 여러개의 컨볼루션 레이어를 쌓으면서 뒤쪽 레이어로 갈수록 필투의 수를 점점 늘리기 때문에 이미지의 마지막 차원 수는 점점 많아진다.      
image.png

컨볼루션 레이어를 생성하는 코드는 다음과 같다
```
conv1 = tf.keras.layers.Conv2D(kernel_size=(3,3), strides=(2,2), padding='vallid',filters=16)
```

주요인수는 kernel_size, strides, padding, filters 네가지이다.      
* kernel_size는 필터 행렬의 크기이다. 이것은 수용 영역(receptive filed)이라고도 부른다. 앞의 숫자는 높이 뒤는 너비, 숫자를 하나만 쓸경우 높이와 너비를 같게한다는 뜻이다      
* strides는 필터가 계산과정에서 한 스텝마다 이동하는 크기이다. 기본값은 (1,1) 이고 (2,2)등으로 설정할 경우 한칸식 건너뛰면서 계산한다. (높이, 너비) 하나만쓸경우 높이, 너비가 같다. strides가 달라지면 결과 이미지의 크기에 영향을 준다.       
image.png

* padding은 컨볼루션 연산 전에 입력 이미지 주변에 빈 값을 넣을지 지정하는 옵션으로 'valid'와 'same'이라는 2가지 옵션중 하나를 사용한다.     
valid는 우리가 봐온 것과 동일 한 방식으로 빈칸을 사용하지 않닌다.     
same은 빈 값을 넣어 출력 이미지의 크기를 입력과 같도록 보존한다      
* filters는 필터의 개수이다. 필터의 개수는 네트워크가 얼마나 많은 특징을 추출할 수 있는지 결정하기 대문에 많을 수록 좋지만, 너무 많으면 느리고 과적합이 발 생할 수도있다.       
가장 유명한 컨볼루션 신경망 네트워크 중 하나인 VGG는 네트워크가 깊어질수록 필터의 수를 2배씩 늘려나간다(64 -> 128 -> 256 -> 512)

### 6.2.2 풀링 레이어

이미지를 구성하는 픽셀 중 인접한 픽셀들은 비슷한 정보르 갖고 있는 경우가 많다. 이런 이미지의 크기를 줄이면서 중요한 정보만 남기기 위해 서브 샘플링이라는 기법을 사용한다. 중요한 정보만 남기는 효율적인 메모리 사용에 도움이 되고 계산할 정보도 줄어들기때문에 과적합을 방지하는 효과도 있다.     
이때 사용되는 레이어가 pooling layer이다

풀링레이어는 Max풀링 Average풀링 등이있다. 이 가운데 컨볼루션 레이어에는 Max 풀링 레이어가 더 많이 쓰인다.
```
pool1 = tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=(2,2))
```
두개의 주요 인수 중 pool_size는 한번에 Max 연산을 수행할 범위이다. pool_size=(2,2)이므로 높이 2 너비 2의 사각형 안에서 최댓값만 남기는 연산을 하게 된다.     
strides는 Conv2D레이어에서 나온 것과 동일하게 계산 과정에서 한 스텝마다 이동하는 크기이다     
image.png

pool_size = (2,2),와 strides=(2,2)옵션으로 MaxPool2D레이어를 사용하면 결과 이미지의 크기는 너비, 높이가 각각 절반으로 줄어든다 홀수라면 내림값이된다(5 -> 2)

### 6.2.3 드롭아웃 레이어


드롭아웃은 네트워크의 과적합을 막기위한 방법 중하나이다      
image.png

드롭아웃은 무작위로 뉴런의 부분집합을 제거하는 것이다. 네트워크가 학습할 때 같은 레이어에 있는 뉴런들은 결과값에 의해 서로 같은 영향을 받는다. 따라서 각 뉴런의 계산 결과를 모두 더해서 나오는 결과값은 한쪽으로 치우치게 된다. 이를 막기 위한 드롭아웃 레이어는 학습 과정에서 확률적으로 일부 뉴런에 대한 연결을 끊고 테스트할때는 정삭적으로 모든 값을 포함해서 계산한다. 
```
pool1 = tf.keras.layers.Dropout(rate=0.3)
```
주요 인수는 rate로 제외할 뉴런의 비율을 나타낸다.    

## 6.3 Fashion MNIST 데이터세트에 적용하기

````
import tensorflow as tf

fashion_mnist = tf.keras.datasets.fashion_mnist
(train_X, train_Y), (test_X, test_Y) = fashion_mnist.load_data()

train_X = train_X/255.0
test_X = test_X/255.0
````

이미지는 보통 채널을 가지고 있고(컬러는 3채널, 흑백은 1채널) Conv2D레이어는 채널을 가진 형태의 데이터를 받도록 기본적으로 설정돼 있기 때문에 데이터가 채널을 갖도록 데이터의 Shape을 바꾼다.

````
# reshape 이전
print(train_X.shape, test_X.shape)

train_X = train_X.reshape(-1, 28, 28, 1)
test_X = test_X.reshape(-1, 28, 28, 1)

# reshape 이후
print(train_X.shape, test_X.shape)
````

데이터를 확인하기 위해 matplotlib.pyplot을 사용해 그래프를 그려보자

````
import matplotlib.pyplot as plt
# 전체 그래프의 크기를 width=10, height = 10으로 지정.
plt.figure(figsize=(10,10))
for c in range(16):
  # 4행 4열로 지정한 그리드에서 c+1번재의 칸에 그래프를 그린다.
  plt.subplot(4,4,c+1)
  plt.imshow(train_X[c].reshape(28,28), cmap='gray')

plt.show()

#훈련데이터의 16번째 까지의 라벨을 출력
print(train_Y[:16])
````

이제 모델을 생성하겠다. 비교를 위해 풀링없이 컨볼루션 레이어만 사용한 모델을 정의해보겠다

````
model = tf.keras.Sequential([
  tf.keras.layers.Conv2D(input_shape=(28,28,1), kernel_size=(3,3), filters=16),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=32),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=64),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(units=128, activation='relu'),
  tf.keras.layers.Dense(units=10, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
````

````
# 느리기 때문에 하드웨어가속기 GPU로 바꾸고 실행
history = model.fit(train_X, train_Y, epochs=25, validation_split=0.25)

import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], 'b-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'r--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()

model.evaluate(test_X, test_Y, verbose=0)
````

그래프를 보면 loss는 감소하고 val_loss는 증가하는 정형적인 과적합 형태를 띈다. 정확도도 검증데이터의 경우 오히려 감소하는 형태이다.     
테스트 데이터의 정확도도 만족할 만한 수준이 아니다.      
이제 풀링레이어와 드롭아웃 레이어를 모두 사용해 보겠다.

````
model = tf.keras.Sequential([
  tf.keras.layers.Conv2D(input_shape=(28,28,1), kernel_size=(3,3), filters=32),
  tf.keras.layers.MaxPool2D(strides=(2,2)),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=64),
  tf.keras.layers.MaxPool2D(strides=(2,2)),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=128),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(units=128, activation='relu'),
  tf.keras.layers.Dropout(rate=0.3),
  tf.keras.layers.Dense(units=10, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
````

풀링레이어가 이미지크기를 줄여주고 있기때문에 파라미터가 줄어들었다.

````
history = model.fit(train_X, train_Y, epochs=25, validation_split=0.25)

import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], 'g-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()

model.evaluate(test_X, test_Y, verbose=0)
````

정확도 89프로 수준으로 뛰어난 모델은 아니다.

## 6.4 퍼포먼스 높이기

퍼포먼스를 높이는데 여러 방법중 대표적이고 쉬운 두가지 방법은 더 많은 레이어 쌓기와 이미지 보강이다.

 ### 6.4.1 더많은 레이어쌓기

딥러닝에서 네트워크 구조를 깊게 쌓는 것이 가능해진 후 컨볼루션네트워크에선 컨볼루션레이어가 중첩된 더 깊은 구조가 계속해서 나타나고 계속 계선되었다.     
VGG의 스타일로 fashion mnist데이터를 분류해보겠다.

````
model = tf.keras.Sequential([
  tf.keras.layers.Conv2D(input_shape=(28,28,1), kernel_size=(3,3), filters=32,
                         padding = "same", activation='relu'),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=64, padding='same', activation='relu'),
  tf.keras.layers.MaxPool2D(pool_size=(2,2)),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=128,padding = "same", activation='relu'),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=256, padding='valid', activation='relu'),
  tf.keras.layers.MaxPool2D(pool_size=(2,2)),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(units=512, activation='relu'),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Dense(units=256, activation='relu'),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Dense(units=10, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
````

vggnet은 여러개의 구조로 실험했는데 그중 19개의 레이어가 겹쳐진 vgg-19가 제일 깊은 구조이다.     
vgg-19는 특징 추출기의 초반에 컨볼루션 레이어를 2개 겹친 뒤 풀링 레이어 1개를 사용하는 패턴을 2차례, 그 후 컨볼루션 레이어를 4개 겹친 뒤 풀링레이어 1개를 사용하는 패턴을 3 차례 반복한다.     
해당코드는 대상이미지가 작고 연산의 한계가 있으니 컨볼루션레이어를 2개 겹치고 풀링 레이어를 1개 사용하는 패턴을 2차례 반복했다. 드롭아웃으로 과적합을 방지했고, flatten레이어 다음의 3개의 Dense사이에도 드롭아웃을 놓았다. vggnet처럼 컨볼루션 레이어와 Dense레이어 수를 세면 vgg-7정도가 된다.

VGG19보다는 깊이가 얕지만 파라미터수는 520만개로 적지않다. 저번모델보다 20배 정도 증가한 숫자이다

````
history = model.fit(train_X, train_Y, epochs=25, validation_split=0.25)

import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], 'g-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()

model.evaluate(test_X, test_Y, verbose=0)
````

val_loss가 증가하지않는 그래프를 얻었다. 아직 과적합되지않아서 에포크수를 늘려도 될 것같다

### 6.4.2 이미지보강

이미지 보강은 훈련데이터에 없는 이미지르 새롭게 만들어내서 훈련데이터를 보강하는것이다. 이때 새로운이미지는 훈련데이터의 이미지를 원본으로 삼고 일정한 변형을 가해 만들어진다.       
이미지를 가로로뒤집거나 신발코가 오른쪽을 향하게 하거나 약간회전, 기울기, 일부확대 를 통해서 다양한 이미지를 만들고 훈련한다

imageDataGenerator는 이미지보강을 쉽게해주는 라이브러리이다.

이미지가 조금씩 변행되었다

ImageDataGenerator의 주요 인수들은 rotation_range, zoom_range, shear_range등이있다.     
flow() 함수는 실제로 보강된 이미지를 생성한다. 이 함수는 iterator객체를 만드는데 이 객체에서는 값을 순차적으로 꺼낼 수 있다. 꺼내는 방법은 next()함수를 사용하는 것이다. 한번에 생성할 이미지의 양을 batch_size로 설정한다.

실제로 훈련데이터 이미지를 보강하기 위해 다량의 이미지를생성하고 학습을 위해 훈련데이터에 추가해보겠다

````
image_generator = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.10,
    shear_range=0.5,
    width_shift_range=0.10,
    height_shift_range=0.10,
    horizontal_flip=True,
    vertical_flip=False)

augment_size = 30000

randidx = np.random.randint(train_X.shape[0], size=augment_size)
x_augmented = train_X[randidx].copy()
y_augmented = train_Y[randidx].copy()
x_augmented = image_generator.flow(x_augmented, np.zeros(augment_size),
                                   batch_size=augment_size,shuffle=False).next()[0]

#원래 데이터인 x_train에 보강된 이미지 추가
train_X = np.concatenate((train_X, x_augmented))
train_Y = np.concatenate((train_Y, y_augmented))

print(train_X.shape)
````

30000개의 이미지를 추가하기위해 augment_size를 30000으로 놓고 변형할 이미지의원본을 뽑기위해 np.random.randint()함수를 써서 0~59999의 범위중 30000개의 정수를 뽑았다. 이때는 중복될 수도있다. 중복이 싫다면 np.random.randint()대신 np.random.choice() 함수를 사용하고 replace인수를 False로 설정하면된다.     
randidx는 [2,134,2455,....]와같은 정수로 된 넘파이array이다. train_X[randidx]로 randidx가 가리키는 이미지를 한번에 선택 가능하다.      
다음으로 ImageDateGenerator의 flow()함수로 30000개의 새로운 이미지를 만든다

````
model = tf.keras.Sequential([
  tf.keras.layers.Conv2D(input_shape=(28,28,1), kernel_size=(3,3), filters=32,
                         padding = "same", activation='relu'),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=64, padding='same', activation='relu'),
  tf.keras.layers.MaxPool2D(pool_size=(2,2)),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=128,padding = "same", activation='relu'),
  tf.keras.layers.Conv2D(kernel_size=(3,3), filters=256, padding='valid', activation='relu'),
  tf.keras.layers.MaxPool2D(pool_size=(2,2)),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(units=512, activation='relu'),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Dense(units=256, activation='relu'),
  tf.keras.layers.Dropout(rate=0.5),
  tf.keras.layers.Dense(units=10, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
````
